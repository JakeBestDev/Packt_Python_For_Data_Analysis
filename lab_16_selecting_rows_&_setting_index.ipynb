### <span style="color:green">Conditional filtering -  boolean indexing </span>.

```python
hero['alignment']
```

```python
hero['alignment'] == 'bad'
```

```python
hero.loc[hero['alignment'] == 'bad']
```

```python
#iloc cannot use the boolean statement. Try a mask. See below
hero.iloc[hero['alignment'] == 'bad']
```

```python
mask = hero['alignment'] == 'bad'
```

```python
mask
```

```python
mask.values
```

```python
## Now works
hero.iloc[mask.values]
```

### Direct Boolean Indexing

```python
hero[hero['alignment'] == 'bad']
```

```python
(hero['alignment'] == 'bad') & (hero['height']>=200)
```

```python
hero[(hero['alignment'] == 'bad') & (hero['height']>=200)]
```

```python
hero[(hero['hair color'] == 'Blond') | (hero['hair color']=='No Hair')]
```

```python
## Same way to do the above 
hero[hero['hair color'].isin(['No Hair','Blond'])]
```

```python
hero[~hero['alignment'].isin(['good','bad'])]
```
