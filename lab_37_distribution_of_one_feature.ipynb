# Distribution Of One Feature

#### A Feature - It is mainly a machine learning term. A feature is an individual property or characteristic, and each of these characteristics are often stored within a column of DataFrame.

##### There are two types of Features:
###### 1) A Numeric Feature
###### 2) A Categorical Feature

#### Each Feature has a set number of ways to analyze them.
##### Numeric Feature:
###### 1) .describe()
###### 2) Histogram
###### 3) Boxplot
###### 4) KDE - kernel density estimation

##### Categorical Feature:
###### 1) .value_counts()
###### 2) Bar Graph (counts)

## Imports and Data

```python
import pandas as pd
```

```python
air_quality=pd.read_pickle('air_quality.pkl')
```

```python
air_quality.columns
```

## Numeric Features

### .describe() method

```python
## Lets look at only numeric columns first.
## These are single numeric features

air_quality[['PM2.5','TEMP','PRES']].describe()
```

### Histogram (hist())

```python
## Now lets use the .hist() method to view these numeric features distrubutions
## Looking at the histogram we can see that:
## PM2.5 has a skewed to the right tail = Meaning most of the data is left leaning
## TEMP has a Bimodal Disrobution = Since there are 2 seperate peaking bars it is Bimodal
## PRES has a Bell Curve

air_quality[['PM2.5','TEMP','PRES']].hist(figsize=(20,10), bins=20)
```

### Boxplot (.boxplot())

```python
## Now lets use the .boxplot() method to view these numeric features distrubutions

air_quality[['PM2.5','TEMP','PRES']].boxplot(figsize=(20,10))
```

```python
## Lets look at the boxplot of a single column
## Notice how using the .boxplot() method doesn't work for a single series or single column

air_quality['TEMP'].boxplot(figsize=(20,10))
```

```python
## Instead of using the .boxplot() method we can use the .plot() method
## The box plot can be used to visualize outliars in a dataset.
## Go back to Lab  27 and 28 to go more indepth

air_quality['TEMP'].plot(kind='box')
```

### KDE Plot

```python
## Now lets look at the KDE plot
## This is a kernel density estimation or kde plot.
## For now, just know that it is a smootherand more complicated version of the histogram based on estimation.

air_quality['TEMP'].plot(kind='kde')
```

```python
## Lets compare the above KDE to a Histogram
## Comparing this histogram to the kde plot, you can see that they show a similar shape,
## While the y axis on the histogram shows the count of the values within each bin,the kde's y axis shows the estimated density.

air_quality['TEMP'].plot(kind='hist')
```

## Categorical Feature

### value_counts() Method

```python
air_quality['PM2.5_category'].value_counts()
```

```python
## use the normalize=True parameter to get the proportions each category makes up of the total
air_quality['PM2.5_category'].value_counts(normalize=True)
```

```python
## Now how do we look at each of the categorical columns using the .value_counts() method?
## A for look of course!
## First we need the names of all the categorical columns

cat_col = air_quality.select_dtypes(include=['object','category','bool']).columns
cat_col
```

```python
## Now the for loop
## This is a handy way to view the value_counts() of each column easily

for columns in cat_col:
    print(air_quality[columns].value_counts())
    print()
```

### Bar Plot (counts)

```python
## A barcount plot shows the categories on the x axis and count for each category on the y axis, 
## so its data should be based on the value_counts method.
## This plot visualizes the count for different categories.

air_quality['PM2.5_category'].value_counts().plot(kind= 'bar')
```

```python
## Now lets try with the TEMP_category
## The reason the TEMP_category is equal across each bar is because in a preious lab we grouped the data into 5 equal bins.
## Look ar Lab 31 binning for this

air_quality['TEMP_category'].value_counts().plot(kind= 'bar')
```
