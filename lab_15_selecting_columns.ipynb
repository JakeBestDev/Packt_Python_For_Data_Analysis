```python
# Select all number columns
hero.select_dtypes(include = 'number')
```

```python
# Select all non-number column types
hero.select_dtypes(exclude = 'number')
```

```python
## Selects only ints and category columns
hero.select_dtypes(include=['int64','category'])
```

```python
hero.select_dtypes(include=['int64','category']).info()
```
