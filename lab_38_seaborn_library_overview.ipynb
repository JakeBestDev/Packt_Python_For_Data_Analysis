## Seaborn Basics: displot()

### Histogram

```python
## Lets start with a histogram
## Use the .displot function. 
## NOTE:: THE .displot function BY DEFAULT sets its 'kind' parameter to 'hist' for Histogram.
## Therefore no need to use the 'kind' parameter for histograms


## Set data to the dataframe you are using and set the x to what column you'd like for the x-axis
## By default with a histogram no need to set a y because it will use record counts

sns.displot(data=air_quality, x='TEMP')
```

```python
## If we wanted we could change the x to a y to rotate the plot about its origin to the y axis while having the count of records on the x-axis
sns.displot(data=air_quality, y='TEMP')
```

```python
## We can set the width of each bin to be a certain amount.
## NOTE: This is not setting the number of bins the data is grouped into but the size of the bin along the x-axis
## Notice how the below bins appear to be increments of 5 for TEMP

sns.displot(data=air_quality, x='TEMP', binwidth = 5)
```

```python
## We can set the number of bins we want to use using the bins parameter
## This now shows 5 total bins that the data is grouped into

sns.displot(data=air_quality, x='TEMP', bins = 5)
```

```python
## We will also want to change the width, height, and aspect ratio between width and height of the plots
## Aspect formula: Aspect * Height = Width
## So for this Width = 9

sns.displot(data=air_quality, x='TEMP', bins = 5, height = 3, aspect = 3)
```

```python
## Aspect change example
## Aspect formula: Aspect * Height = Width
## So for this Width = 1.5

sns.displot(data=air_quality, x='TEMP', bins = 5, height = 3, aspect = 0.5)
```

### KDE

```python
## To set a kde use the 'kind' parameter and set it to 'kde'
## Notice, as mentioned above, there was no need to use the kind parameter for histogram as it is default

sns.displot(data=air_quality, x='TEMP',kind='kde')
```

```python
## You can apply the kde over the histogram using the kde parameter and setting it to True
## Since the kind parameter is still default to a histogram that is why it appears

sns.displot(data=air_quality, x='TEMP',kde=True)
```

```python
## We can add a rug plot to the histogram using the rug parameter
## A rug plot shows each observation with tick marks on the x axis.It is also used to visualize the distribution of data. 
## It shows one tick mark for eachdata at that value, so it's showing the exact location of individual observations.
## Rug plots are only really helpful wwith smaller datasets. This dataset is far too big to be helpful as you'll just see a large number of ticks at the bottom

sns.displot(data=air_quality, x='TEMP',kde=True, rug = True)
```

## Seaborn Basics: catplot()

### Boxplot

```python
## To build a boxplot use the catplot function
## Set the kind parameter to be 'box'
## We set the y to be the column because it's a bit easier to read

sns.catplot(data=air_quality, y='TEMP',kind='box')
```

```python
## We can set the TEMP to be on the x-axis if needed 

sns.catplot(data=air_quality, x='TEMP',kind='box')
```

### Bar (Count) Plot

```python
## For categorical data we use a Bar (count) plot
## Set the kind parameter to be 'count'
## Notice how the x-axis lables are hard to read. 

sns.catplot(data=air_quality, x='PM2.5_category',kind='count')
```

```python
## use the aspect parameter to change the size of the plot
## MUCH EASIER TO READ NOW!!

sns.catplot(data=air_quality, x='PM2.5_category',kind='count', aspect=2.5)
```
