## Sorting Data

```python
import pandas as pd
hero = pd.read_pickle('hero.pkl')
```

```python
## Using previous data from lab 18 & 19
hero.loc[hero['race'] == '-','race'] = 'Unknown'
hero.loc[10000]=hero.loc[533]
hero.iloc[536] = hero.iloc[533]
hero['height'] = hero['height']*0.0328084
hero['bmi'] = hero['weight']/(hero['height']/3.28084)**2
hero_ref = hero
hero_copy= hero.copy()
hero['original_ind']=True
hero_ref['ref_ind']=True
hero_copy['copy_ind']=True
```

### .sort_values

```python
# by default its in ascending order
hero.sort_values(by='race')
```

```python
hero.sort_values(by='race',ascending =False)
```

```python
hero.sort_values(by = 'bmi',ascending=False).head()
```

```python
hero[['race','height']].sort_values(by=['race','height'])
```

```python
# Notice it is now sorted by height primarily then by race compared to above
hero[['race','height']].sort_values(by=['height','race'])
```

```python
## Height is primary from shortest to tallest (ascending) and race is then the secpondary going in descending order
hero[['race','height']].sort_values(by=['height','race'],ascending=[True,False])
```

### .nlargest & .nsmallest

```python
hero.nlargest(n=10,columns= 'height')
```

```python
## Same as above using sort values
hero.sort_values('height',ascending=False).head(10)
```
