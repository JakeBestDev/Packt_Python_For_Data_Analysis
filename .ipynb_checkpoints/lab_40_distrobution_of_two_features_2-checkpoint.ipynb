# Distrobution of Two Features - Part 2

### For two numericals, we could do scatterplot using different Seaborn functions including: 
#### - relplot(), pairplot(), and jointplot(). We can also use correlation matrix to do a heatmap.

### For one categorical and one numerical, we could do:
#### - Histograms for different categories usingthe Seaborn function, displot(), or a line plot using the function relplot() or any of these plots including bar/box/boxen/strip/swarm/ and violin plots with the function catplot.

### For two categorical features, we could do:
#### - A count plot for differentcategories using the function catplot.

## Data and Imports

```python
import pandas as pd
import seaborn as sns
```

```python
air_quality = pd.read_pickle('air_quality.pkl')
```

```python
air_quality.columns
```

## One Categorical and One Numerical Feature

### Histogram

```python
## As we recall from previous lessons, the displot function creates a histogram by default
## By default as well the histogram is created with a single feature and its counts 
## To add a numeric feature and group it by a categorical feature we much use the 'hue' or 'col' paramater and assign it the categorical feature
## See below:

sns.displot(data=air_quality,x='TEMP', hue='quarter')
```

```python
## Now lets try the 'col' parameter instead of 'hue'
## Notice how the histogram is now broken into 4 different graphs based on the categorical data

sns.displot(data=air_quality,x='TEMP', col='quarter')
```

### Line Chart

```python
## We can also use a line chart to view the relationship between a numeric and categorical feature
## Compared to the histogram we use the relplot() function.
## You must set the x and y of the line chart and then set the kind parameters to 'line'

## You may be wondering how seaborn determined the temps for each quarter since there are so many. See next cell

sns.relplot(data=air_quality, x='quarter',y='TEMP', kind='line')
```

```python
## You may be wondering how seaborn determined the temps for each quarter since there are so many. See below:

## We can use the groupby method to group the data by the quarter and then compare the mean TEMP of each quarter.
## Notice how the mean TEMPS of each quarter match that of the above line chart.

air_quality.groupby('quarter')['TEMP'].mean()
```

### Bar Chart

```python
## we can also easily create a barchat using the catplot() method
## We set the aspect = 2 to make the chart easier to read
## Notice how similar it looks to the line chart! It takes the mean of numerical 'TEMP' feature and plots it across the categorical feature, quarter

sns.catplot(data=air_quality,x='quarter',y='TEMP', kind='bar', aspect=2)
```

### catplot(kind=) Many different kinds!

##### There are a lot of options available through seaborn!
##### catplot is dedicated to categorical features! 
##### We can set the 'kind' parameter to be: "strip", "swarm", "box", "violin", "boxen", "point", "bar", or "count".
##### Lets try a few!

#### Box

```python
## Setting kind to 'box'
## Notice how each quarter is now broken down into a seperate box plot

sns.catplot(data=air_quality,x='quarter',y='TEMP', kind='box', aspect=2)
```

#### Boxen

```python
## Setting kind to 'boxen'
## The boxen plot draws a plot that is similar to a box plot, 
## but optimized for showing more information about the shape of the distribution.

## It is more suited for larger datasets. 
## As you can see,for example, the quarter 1 boxen plot has thesame box in the middle shown on the box plot, 
## which has information about the quartiles, but there are also these extra boxes showing the distribution outside the quartiles.

sns.catplot(data=air_quality,x='quarter',y='TEMP', kind='boxen', aspect=2)
```

#### Strip

```python
## Setting kind to 'strip'
## The strip plot is basically scatterplot, but with one of the features being categorical, such as 'quarter'. 
## The strip plot is better for smaller datasets! As you can see there isn't much we can draw from these scatterplots outside of viewing outliars

sns.catplot(data=air_quality,x='quarter',y='TEMP', kind='strip', aspect=2)
```

#### swarm

```python
## Setting kind to 'swarm'
## Similar to strip, swarm is better for smaller datasets
## It will also take too long if we try to use the whole air_quality dataset. Lets filter by the PM2.5 > 300

## As you can see, the swarm plot, similar toa strip plot, provides a visualization to see the spread of values of TEMP along the y axis.
## The difference between the strip plot and the swarm plot is that the swarm plot spreadsout the data points to avoid overlap.
## So you can see that for strip plot, these points are all clustered together, 
## while for swarm plot, you can see roughly how many points are for, say, TEMP of 25 in quarter 2.

## But don't miss this warning messages!!. 
## It says 20.5%, 34.0%, 30.0%, 41.0% of the plots cannot be placed; 
## you may want to decrease the sizeof the markers or use stripplot.That is because the swarm plottries to spread out the points. 
## When there are too many data points ofsimilar values, it can't spread everything nicely.

sns.catplot(data=air_quality[air_quality['PM2.5']>300],x='quarter',y='TEMP', kind='swarm', aspect=2)
```

```python
## Lets reduce the data points even more by filtering PM2.5 > 600 rather than 300
## Notice how the swarm plot only includes data from quarters 1 and 4?
## These are the only quarters with a PM2.5 over 600

## There are also no more warning messages regarding points that could not be included

sns.catplot(data=air_quality[air_quality['PM2.5']>600],x='quarter',y='TEMP', kind='swarm', aspect=2)
```

#### violin

```python
## Setting kind to 'violin'

## The violin plot is a combination of box plot and kernel density estimate (KDE). 
## It is like box plot because it's still showing the distribution of the TEMP for each category of quarters. 
## But instead of only box and whiskers, this functionshows the kernel density estimate curve along the y axis. 
## Additionally, it's a little hard to see, but the quartile and whisker values from the box plotare shown inside the violin.

## For example, we can see that in quarter 1 the TEMP has a peak around 0, with a large range from below -10 to above 20.
## This looks fancy, but in reality we use box plots more often because they are more straightforward.

sns.catplot(data=air_quality,x='quarter',y='TEMP', kind='violin', aspect=2)
```

## Two Categorical Features

### Count Plot

```python
## We will use the cateplot() function as it is best suited for categorical features
## We we recall the count plot can be used for one categorical feature

## To have a count plot work for two categorical features we must use the hue or col parameters. See next cell

sns.catplot(data=air_quality,x='PM2.5_category', kind='count', aspect=3)
```

```python
## To have a count plot work for two categorical features we must use the hue or col parameters. See below:

## Lets try the 'hue' parameter

sns.catplot(data=air_quality,x='PM2.5_category', kind='count', hue ='quarter', aspect=3)
```

```python
## To have a count plot work for two categorical features we must use the hue or col parameters. See below:

## Lets try the 'col' parameter

## It's very hard to read to lets use the col_wrap parameter to view each quarters count chart easier

sns.catplot(data=air_quality,x='PM2.5_category', kind='count', col='quarter', col_wrap=2, aspect=2)
```
