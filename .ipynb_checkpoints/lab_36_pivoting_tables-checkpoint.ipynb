## Pivot Table

### Pivot Table Basics

```python
## This is how to create a basic Pivot Table pd.pivot_table(student, index='sex') 
## For some reason the above basic table throws an error if the 'values' columns that we're not specified
## By default gives the mean of the selected value columns

pd.pivot_table(student, 
               index='sex', 
               values = ['score','age'])
```

```python
## Can use multiple index's 
pd.pivot_table(student, 
               index=['sex', 'internet'], 
               values = ['score','age'])
```

```python
## Can speciify which columns you want as values for the index columns using 'values'

pd.pivot_table(student, 
               index=['sex', 'internet'], 
               values = 'score')
```

### Aggregate by other stats using pivot tables

```python
## use the agg parameter to change how the pivot table value parameters are aggregated

pd.pivot_table(student, 
               index=['sex', 'internet'], 
               values='score',
              aggfunc='median')
```

```python
pd.pivot_table(student, 
               index=['sex', 'internet'], 
               values='score',
              aggfunc='mean')
```

```python
## Can also give a list of functions

pd.pivot_table(student, 
               index=['sex', 'internet'], 
               values='score',
              aggfunc=['min','mean','max'])
```

### Columns Parameter 

```python
## We can further divide the values of the pivot table by specific columns
## In this example we specify studytime in the columns parameter to have the data in the pivot table to divided into those categories
## The NaN in the 4th column for men with no internet who study more than 10 hours indicates we do not have this data 

pd.pivot_table(student, 
               index=['sex', 'internet'], 
               values='score',
              aggfunc='mean',
              columns = 'studytime')
```

### fill_value Parameter

```python
## Notice The NaN in the 4th column for men with no internet who study more than 10 hours indicates we do not have this data 
## We will use the fill_value pivot table function to fill the missing data with a value we determine

pd.pivot_table(student, 
               index=['sex', 'internet'], 
               values='score',
              aggfunc='mean',
              columns='studytime',
              fill_value='-999')
```

### Multiple 'values' Columns

```python
## Notice the data is broken into two seperate halves.
## The Left side is now 'age' which is further broken down by 'studytime' for both 'internet', and 'sex'
## The Right side is now 'scorre' which is further broken down by 'studytime' for both 'internet', and 'sex'

pd.pivot_table(student, 
               index=['sex', 'internet'], 
               values=['score','age'],
              aggfunc='mean',
              columns='studytime',
              fill_value='-999')
```

```python
## We can further madify the above by moving 'internet' from the 'index' to the 'columns' parameter
## Also make 'values' parameter just look for 'score' to simplify

pd.pivot_table(student, 
               index='sex', 
               values='score',
              aggfunc='mean',
              columns=['internet','studytime'],
              fill_value='-999')
```

### Example 1)

```python
## Create a pivot table of the mean of 'score' based on the family size and sex of the student
pd.pivot_table(student,
               index='familysize',
               values='score',
               aggfunc='mean',
               columns='sex')
```

### margins Parameter

```python
## Use the margins parameter to view the total averages of the data

pd.pivot_table(student,
               index='familysize',
               values='score',
               aggfunc='mean',
               columns='sex',
              margins=True)
```

```python
## You can rename the margins using the margins_name

pd.pivot_table(student,
               index='familysize',
               values='score',
               aggfunc='mean',
               columns='sex',
              margins=True,
              margins_name= 'Average Score Total')
```

## Further Practice and Examples

### Working More With aggfunc

```python
## The aggfunc works similar to .agg method where you can further break down what functions you want specific columns to be broken down by.
## Use a dictonary like we used in the .agg method to specify the stats functions we want used for specific columns
## So for example here the columns in the 'values' parameter are used for the aggfunc parameter to specify that we want the min/max of age and the median score

pd.pivot_table(student,
               index='studytime',
               values=['age','score'],
               aggfunc={'age':['max','min'],'score':'median'},
               columns='sex')
```

```python
## Now lets try with a function!

def range_func(series):
    return series.max()-series.min()

pd.pivot_table(student,
               index='studytime',
               values='score',
               columns='sex',
               aggfunc=range_func)
```

```python
## Can also use a lambda function for the above

pd.pivot_table(student,
               index='studytime',
               values='score',
               columns='sex',
               aggfunc=(lambda series:series.max()-series.min()))
```
