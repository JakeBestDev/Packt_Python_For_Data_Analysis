## Subsetting Rows and Columns

```python
import pandas as pd
```

```python
hero = pd.read_pickle('hero.pkl')
```

```python
hero.head(5)
```

```python
#null slice, gets all columns. Same as this hero.loc[0]
hero.loc[0,:]
```

```python
hero.loc[0,'race']
```

```python
hero.iloc[0,3]
```

```python
#hero.loc[2:4,'race':'height'] also works
hero.loc[2:4,['race','hair color','height']]
```

```python
hero.iloc[2:5,3:6]
```

```python
hero.loc[10:20,['hero names','super strength','flight','super speed']]
```

```python
hero.iloc[10:21,[9,27,18,40]]
```

```python
hero.loc[hero['flight'] == True,'name']
```

```python
mask = hero['flight'] == True
hero.iloc[mask.values,0]
```

```python
## This works to filter flight but to get all columns. You can't easily just get the 'name' column in this single statement, unlike above with iloc and loc
hero[hero['flight']==True]
```

```python
# Same as hero['name']
hero.loc[:,'name']
```
