# Distrobution of Multiple Features

#### So to visualize more than two featuresis similar to what we've done. The main idea is to add different parameters to the functions that we've used.

##### We can use the hue parameterto color the plot based on a feature.
##### We can use col to define subsetsto plot separately based on a feature
##### We can also use x and/or y to specifyextra features to show on the x or y axis.

## Imports and Data

```python
import pandas as pd
import seaborn as sns
```

```python
air_quality = pd.read_pickle('air_quality.pkl')
```

```python
air_quality.columns
```

## Multiple Features

```python
air_quality.columns
```

### For this first example we want;
#### 1) a histogram for TEMP
#### 2) With the color assigned to quarter
#### 3) And a seperate histogram for each station

```python
## See below plot:

sns.displot(data=air_quality,x='TEMP', hue = 'quarter' ,col='station')
```

### For this first example we want:
#### 1) A boxplot for TEMP
#### 2) Boxplot categories seperated by quarter 
#### 3) With the color assigned to prior_2016_ind

```python
## See below plot:

sns.catplot(data=air_quality,x='quarter', y='TEMP', hue = 'prior_2016_ind',kind = 'box')
```

```python
## We can also change the box plot to a bar plot thanks to catplot! 

sns.catplot(data=air_quality,x='quarter', y='TEMP', hue = 'prior_2016_ind',kind = 'bar')
```

### For this first example we want:
#### 1) PM2.5_category as the categories
#### 2) A stripplot based on TEMP
#### 3) With the color assigned to quarter

```python
## See below plot:

## No need to set kind parameter as 'strip' because that is the default but I did so anyway.

sns.catplot(data=air_quality,x='PM2.5_category',y='TEMP', hue = 'quarter', kind='strip', aspect = 2)
```

### Pairplot

```python
## Now lets look at a pairplot for 'TEMP', 'PRES', 'PM2.5', 'quarter'

sns.pairplot(data=air_quality[['TEMP', 'PRES', 'PM2.5', 'quarter']],hue = 'quarter')
```

### relplot

```python
## We can create the scatterplots of the above plots using the relplot function

sns.relplot(data=air_quality,x='TEMP',y='PRES', hue='quarter')
```

```python
## We can further the analysis of the plot using the col parameter in the above relplot
sns.relplot(data=air_quality,x='TEMP',y='PRES', hue='quarter', col = 'year', col_wrap=3)
```

```python
## Besides hue and col there are other parameters we can add!

## Like:
## style  which changes the types of dots for readability! We set the quarter to be the different styles of dots 

sns.relplot(data=air_quality,x='TEMP',y='PRES', hue='quarter', col = 'year', col_wrap=3, style='quarter')
```

```python
## Besides hue and col there are other parameters we can add!

## Like:
## size  which changes the size of a specific featue. Notice how the larger the PM2.5 value the larger the dot is going to be
## This indicates to us that the larger the dot the worse the air quality

sns.relplot(data=air_quality,x='TEMP',y='PRES', hue='quarter', col = 'year', col_wrap=3, size='PM2.5')
```
