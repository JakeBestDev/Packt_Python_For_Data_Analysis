## Modifying Values In A Dataframe

```python
import pandas as pd
```

```python
hero  = pd.read_pickle('hero.pkl')
```

```python
# Notice dashes in some columns
hero.head(5)
```

### Change Columns

```python
hero.loc[2,'race']
```

```python
hero.loc[2,'race'] = 'Unknown'
```

```python
# Alan Scott's race is now 'Unknown'
hero.head()
```

```python
hero.iloc[2,3] = '-'
```

```python
# Changed back to - 
hero.head()
```

### Change Whole Column

```python
hero.loc[(hero['race'] == '-'),'race']
```

### Change column

```python
# This is in cm, we want inches
hero['height']
```

```python
hero['height']*0.0328084
```

```python
hero['height'] = hero['height']*0.0328084
```

```python
hero['height']
```

### Change all columns with numbers in dataframe

```python
# Same as hero[['height','weight']] but there could be more numeric columns so a lot of typing hence select_dtypes
hero.select_dtypes(include='number')
```

```python
numeric_cols = hero.select_dtypes(include='number').columns
```

```python
numeric_cols
```

```python
## Now with the numeric_cols above we have easy access to the numeric columns
hero[numeric_cols]
```

```python
## Can now multiply by 100 or do as desired to numeric columns
hero[numeric_cols] = hero[numeric_cols]*100
```

```python
hero[numeric_cols]
```

```python
# Change back
hero[numeric_cols] = hero[numeric_cols]/100
```

```python
hero[numeric_cols]
```
