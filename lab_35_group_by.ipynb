# groupby()

#### Group by is a process involving 3 steps:
##### 1) Split the data into groups based on criteria
##### 2) Apply a function to each group independently
##### 3) Cpmbine the results into a data structure

## Imports and data

```python
import pandas as pd
```

```python
air_quality = pd.read_pickle('air_quality.pkl')
```

```python
air_quality.head()
```

## groupby() Basics

```python
## In this example we are grouping the air_quality dataframe by PM2.5_category
## BUT does not display anything outside of the notification that they are grouped.
## See next cell
air_quality.groupby(by='PM2.5_category').
```

```python
## using .group we can see a dictonary whos keys are the categories and the values are the rows axis lables they apply to
air_quality.groupby(by='PM2.5_category').groups
```

```python
## We can prove the above by using the .head() method
## Notice rows 0-15 are "good" but 16, 18 and 19 are "moderate" this is backed up above by looking at the axis's listed
air_quality['PM2.5_category'].head(20)
```

```python
## We can view all the group names by using the .keys() method
air_quality.groupby(by='PM2.5_category').groups.keys()
```

```python
## We can view specifiec groups by using .get_group() method
air_quality.groupby(by='PM2.5_category').get_group('Good')
```

## Stats Methods and groupby()

### Sorting groups

```python
## We can apply groupby() after sorting data
air_quality.sort_values('date_time').groupby(by = 'year')
```

```python
## We can then use the first() method to get the first records in each group. 
## Basically the earliest records in each year since their sorted
air_quality.sort_values('date_time').groupby(by = 'year').first()
```

```python
## We can then use the last() method to get the last records in each group. 
## Basically the latest records in each year since their sorted
air_quality.sort_values('date_time').groupby(by = 'year').last()
```

### Sizes of Groups

```python
## We can get the sizes of each group using the .size() method
air_quality.groupby(by = 'TEMP_category').size()
```

```python
## The above is similar to the .value_counts() method
air_quality['TEMP_category'].value_counts(sort=False)
```

### Using .groupby() and .agg()

```python
## Lets get the mean TEMP and PM2.5 and group them by quarter
## We can draw from this that the air quality goes down (highter PM2.5) when the TEMP is lower
air_quality[['PM2.5','TEMP','quarter']].groupby(by = 'quarter').mean()
```

```python
## This is the same as the above but a different way you can write it
air_quality.groupby('quarter')[['PM2.5','TEMP']].mean()
```

```python
## We can use differnt methods like describe()
air_quality.groupby('quarter')[['PM2.5','TEMP']].describe()
```

```python
## We can also use the .agg()
air_quality.groupby('quarter')[['PM2.5','TEMP']].agg(['min','max'])
```

```python
## We can make more complex .agg() methods to filter our data
air_quality.groupby('day_of_week_name')[['PM2.5','TEMP','RAIN']].agg({'PM2.5':['min','max','mean'],'TEMP':'mean','RAIN':'mean'})
```

## groupby() Multiple Columns

```python
## We can group by multiple columns and apply statistical methods like mean()
air_quality.groupby(['quarter','TEMP_category'])[['PM2.5','TEMP']].mean()
```

```python
## The order of the .groupby() arguments matter. 
## Lets switch 'quarter','TEMP_category'
air_quality.groupby(['TEMP_category','quarter'])[['PM2.5','TEMP']].mean()
```

```python
### Now lets try with 3 columns
air_quality.groupby(['year','quarter','month'])[[]]
```
