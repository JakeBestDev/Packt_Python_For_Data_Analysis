## Handling Outliers Pt.1

```python
import pandas as pd
wdi_imp = pd.read_pickle('wdi_imp.pkl')
```

### Identifying Outliers

```python
# Don't forget we use .T to transpose the dataframe to see the data easier since the df is so large
# The min, 25%, 50%, 75%, and max are referred to as the 5 number summary in statistics. They give us insights into the columns
# We can spot outliers using describe easily like for example population.
# Population has a max of 7.591945e+09 or 7.59 billion, which is impossible given the population is estimated to be 8.1 billion. 
# So no single country can have that many making it an outlier
wdi_imp.describe().T
```

### Visualizations
##### Visualizations are great ways to spot outliers

## Histographs

```python
## Kinda small so lets enlarge the images
wdi_imp.hist()
```

```python
## Much better! Zoom out so you can see them all
## The y axis is the count of records and the x axis is the number in that particular record for that column.
# So the higher the bar the more records there are with a similar number.
# The further to the right a bar is the larger the value.
# Notice population again. Its got 350+ records on the left of the graph and then smaller, hard to see graphs on the right. Likley due to outliers
# Now notice life_expectancy, its a much more even distrobution, so likley less outliers.
wdi_imp.hist(figsize=(20,20))
```

### There are 3 columns we suspect to have outliers. population, population_density, and atms_per_100000

```python
## We can view the histographs of individual columns like this.
## Issue is its hard to tell how spread out the data is.
wdi_imp['population'].hist()
```

```python
## Adding the bins = 50 to help us see how spread out the data is in this graph. By default it's 10
# Now we can see how spread out the data is around 5-7
wdi_imp['population'].hist(bins = 50)
```

```python
# Lets view the actual data. Using the sort_values we learned earlier we can sort the data by the population in decending order, 
# giving us the largest data first with their country names.

# Take a look! We have some irrelevent data here, like 'world' and 'IDA & IBRD total'
# There is also an issue we can spot at the lower half of the data.
# Doubles od tonga and St. Vincent and the Grenadines.
# Lets look at the columns for a solution
wdi_imp[['country_name','population']].sort_values(by='population', ascending = False)
```

```python
# Looking at the column we can see 'country_names' and 'is_region'. 
# The 'is_region' could be why we're seeing 2 of the same names but different populations.
# Lets see if we can distinguish and confirm this theory

wdi_imp.columns
```

```python
# The 'is_region' column takes a 1 or 0 basically saying its a region or not using bool
# We can filter by that to see how many there are
# Yes there are regions included in the dataset
# So, if we just wanted the country data we'd need to filter out these regions.
wdi_imp.loc[wdi_imp['is_region']==1, 'country_name'].value_counts()
```

```python
# Now lets filter out all region data
wdi_imp = wdi_imp[wdi_imp['is_region'] == 0]
```

```python
# check the shape to see how many were taken. The df was 400+ originally
wdi_imp.shape
```

```python
# The data is still scewed but there is far less data in the middle meaning a large number of outliers were handled.
## We know that the largest population listed that appears to be an outlier is china at 1.4 billion population
# Thus making it not an outlier
# See below to confirm
wdi_imp['population'].hist(bins = 50)
```
